```python
from ultralytics import YOLO
import torch

def train_model():

    print("Loading YOLOv8 model...")
    model = YOLO("yolov8s.pt")


    model.train(
        data="/kaggle/input/datasets/duongvankhanh/fruit-detection/Fruits-detection/fruit.yaml",
        epochs=100,
        imgsz=640,
        batch=16,
        device=0,
        project="runs",
        name="fruit_detector"
    )

    print("Training finished!")

if __name__ == "__main__":
    train_model()
```
